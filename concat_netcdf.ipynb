```python
#!bin/bash/python3

import xarray
from dask.distributed import Client, LocalCluster
import xarray as xr
import glob
```

```python
input_files = glob.glob('/home/davidbyrne/disks/ssd200/wetbulb/analysis/annual_maxima/1_5deg/*')
output_file = '/home/davidbyrne/disks/ssd200/wetbulb/analysis/annual_maxima/annual_maxima_1_5deg.nc'
dimension='year'
```

```python
client = Client("tcp://127.0.0.1:34925")
```

```python
datasets = [ xr.open_dataset(fn, chunks={}) for fn in input_files ]
datasets_concat = xr.concat(datasets, dim=dimension)
datasets_concat = datasets_concat.chunk({dimension:-1})
datasets_concat = datasets_concat
datasets_concat.to_netcdf( output_file )
```
