```python
import dask
from dask import delayed
import matplotlib.pyplot as plt
import numpy as np
from dask.distributed import Client, LocalCluster
import xarray as xr
import os
import pyextremes as pe
import sys
import dask.array as da

client = Client("tcp://127.0.0.1:44177")
client.scheduler_info()
```

```python
# Create random array
array = da.random.random((10000, 200, 200), chunks=(-1, 10, 10))
```

```python
%%time
# 1. Just use dask.array.max
max1 = da.max(array, axis=0).compute()
```

```python
%%time
def loop_max(chunk):
    return da.max(chunk, 0)

max2 = da.map_blocks(loop_max, array, drop_axis=0, )
max2 = max2.compute()
```

```python
# 3. Via delayed objects
chunks = array.to_delayed().ravel()

@dask.delayed
def test_function(chunk):
    mm = da.max(chunk, axis=0)
    return mm
```

```python
%%time
results = [test_function(b) for b in chunks] 
#results = [rr.compute() for rr in results]
results2 = client.compute(results)
```
